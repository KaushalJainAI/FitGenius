# 03 - Collaborative Filtering
**Unit 3: Matrix Factorization and Latent Factors**

Conducting SVD on User-Recommendation interactions to find latent fitness concepts.

## Overview
This notebook implements User-Based and Item-Based collaborative filtering using
Matrix Factorization (Truncated SVD) to unearth hidden relationships between users
and specific workout/diet regimens.


```python
import pandas as pd
import numpy as np
import os
from sklearn.decomposition import TruncatedSVD
from sklearn.model_selection import train_test_split

NOTEBOOK_DIR = os.path.dirname(os.path.abspath('__file__'))
DATA_PATH = os.path.join(NOTEBOOK_DIR, 'data', 'merged_fitness_data.csv')

if not os.path.exists(DATA_PATH):
    print(f"ERROR: {DATA_PATH} not found. Run merge_datasets.py first.")
else:
    df = pd.read_csv(DATA_PATH)
    df = df[df['diet_recommendation'].notna()].copy()
    
    train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
    
    interaction_matrix = pd.crosstab(train_df.index, train_df['diet_recommendation'])
    print(f"Interaction matrix shape: {interaction_matrix.shape}")
    print(f"Unique users: {interaction_matrix.shape[0]}, Unique plans: {interaction_matrix.shape[1]}")
    
    if not interaction_matrix.empty:
        n_components = min(10, interaction_matrix.shape[1]-1)
        svd = TruncatedSVD(n_components=n_components, random_state=42)
        latent_space = svd.fit_transform(interaction_matrix)
        
        print(f"✓ Latent space shape: {latent_space.shape}")
        print(f"✓ Modeled {len(interaction_matrix.columns)} plans into {n_components} latent concepts.")
        print(f"✓ Explained variance ratio: {svd.explained_variance_ratio_.sum():.4f}")
        
        reconstructed = svd.inverse_transform(latent_space)
        reconstruction_error = np.sqrt(np.mean((interaction_matrix.values - reconstructed) ** 2))
        print(f"✓ Reconstruction error (RMSE): {reconstruction_error:.4f}")
    
    print("\n✓ Collaborative filtering model ready for hybrid integration.")
```

## Latent Concept Interpretation

Each latent factor represents a hidden pattern in user preferences.
These could correspond to:
- User fitness levels (beginner vs advanced)
- Health conditions (diabetic vs healthy)
- Dietary restrictions (vegan vs non-veg)
- Workout preferences (cardio vs strength)
