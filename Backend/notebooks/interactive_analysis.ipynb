# Interactive Data Analysis with Plotly Express

This notebook provides dynamic, interactive visualizations of all fitness datasets.

**Features:**
- Interactive hover tooltips
- Zoom, pan, and zoom-to-fit capabilities
- Dynamic filtering and updates
- Multiple chart types for comprehensive analysis

```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')

import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

pd.set_option('display.max_columns', None)

sources = {
    'Gym Recommendation': 'data/gym_recommendation.csv',
    'Diet Recommendations': 'data/diet_recommendations.csv',
    'Personalized Medical Diet': 'data/personalized_medical_diet.csv',
    'Diet Workout Dataset': 'data/diet_workout_dataset.csv'
}
MERGED_PATH = 'data/merged_fitness_data.csv'
```

## 1. Dataset Overview

```python
info_data = []
for name, path in sources.items():
    df = pd.read_csv(path)
    info_data.append({
        'Dataset': name,
        'Rows': len(df),
        'Columns': len(df.columns),
        'Missing': df.isnull().sum().sum(),
        'Duplicates': df.duplicated().sum()
    })

info_df = pd.DataFrame(info_data)

fig = px.bar(
    info_df, 
    x='Dataset', 
    y='Rows',
    color='Dataset',
    title='Dataset Size Comparison',
    text='Rows',
    hover_data=['Columns', 'Missing', 'Duplicates']
)
fig.update_layout(showlegend=False, height=500)
fig.show()
```

## 2. Missing Values Analysis

```python
for name, path in sources.items():
    df = pd.read_csv(path)
    missing = df.isnull().sum()
    missing = missing[missing > 0].sort_values(ascending=True)
    
    if len(missing) > 0:
        fig = px.bar(
            x=missing.values,
            y=missing.index,
            orientation='h',
            title=f'{name} - Missing Values',
            labels={'x': 'Missing Count', 'y': 'Column'},
            text=missing.values,
            color=missing.values,
            color_continuous_scale='Reds'
        )
        fig.update_layout(height=400, showlegend=False)
        fig.show()
    else:
        print(f'{name}: No missing values')
        print('-' * 50)
```

## 4. DIET RECOMMENDATIONS Dataset - Interactive Analysis

```python
diet_df = pd.read_csv('data/diet_recommendations.csv')

fig = make_subplots(rows=2, cols=2,
                   subplot_titles=('Age', 'BMI', 'Daily Caloric Intake', 'Weekly Exercise Hours'),
                   horizontal_spacing=0.12, vertical_spacing=0.15)

fig.add_trace(go.Histogram(x=diet_df['Age'], nbinsx=30, name='Age', marker_color='coral'), row=1, col=1)
fig.add_trace(go.Histogram(x=diet_df['BMI'], nbinsx=30, name='BMI', marker_color='forestgreen'), row=1, col=2)
fig.add_trace(go.Histogram(x=diet_df['Daily_Caloric_Intake'], nbinsx=30, name='Calories', marker_color='goldenrod'), row=2, col=1)
fig.add_trace(go.Histogram(x=diet_df['Weekly_Exercise_Hours'], nbinsx=30, name='Exercise', marker_color='steelblue'), row=2, col=2)

fig.update_layout(title_text='Diet Dataset - Distribution Overview', height=700, showlegend=False)
fig.show()
```

```python
fig = px.pie(
    diet_df, 
    names='Gender', 
    title='Gender Distribution',
    hole=0.5,
    color_discrete_sequence=['#ff9999', '#66b3ff']
)
fig.show()
```

```python
fig = px.bar(
    diet_df, 
    y='Disease_Type', 
    title='Disease Types Distribution',
    orientation='h',
    color='Disease_Type',
    color_discrete_sequence=px.colors.qualitative.Set3
)
fig.update_layout(showlegend=False)
fig.show()
```

```python
fig = px.box(
    diet_df, 
    x='Severity', 
    y='Daily_Caloric_Intake', 
    title='Caloric Intake by Disease Severity',
    color='Severity',
    points='outliers'
)
fig.update_layout(showlegend=False)
fig.show()
```

```python
fig = px.scatter_matrix(
    diet_df[['Age', 'BMI', 'Daily_Caloric_Intake', 'Cholesterol_mg/dL', 'Glucose_mg/dL']].dropna(),
    dimensions=['Age', 'BMI', 'Daily_Caloric_Intake', 'Cholesterol_mg/dL', 'Glucose_mg/dL'],
    title='Diet Dataset - Feature Correlations',
    color='Severity'
)
fig.update_layout(height=800)
fig.show()
```

## 6. DIET WORKOUT DATASET - Interactive Analysis

```python
workout_df = pd.read_csv('data/diet_workout_dataset.csv')

fig = make_subplots(rows=1, cols=3,
                   subplot_titles=('Heart Rate', 'Blood Oxygen Level', 'Body Temperature'))

fig.add_trace(go.Histogram(x=workout_df['HeartRate'], nbinsx=30, name='HeartRate', marker_color='crimson'), row=1, col=1)
fig.add_trace(go.Histogram(x=workout_df['BloodOxygenLevel'], nbinsx=30, name='BloodO2', marker_color='royalblue'), row=1, col=2)
fig.add_trace(go.Histogram(x=workout_df['BodyTemperature'], nbinsx=30, name='Temp', marker_color='darkorange'), row=1, col=3)

fig.update_layout(title_text='Workout Dataset - Physiological Metrics', height=400, showlegend=False)
fig.show()
```

```python
fig = px.scatter(
    workout_df, 
    x='HeartRate', 
    y='BodyTemperature', 
    color='BloodOxygenLevel',
    title='Heart Rate vs Body Temperature (colored by Blood Oxygen)',
    color_continuous_scale='RdYlBu',
    hover_data=['ExercisePaln', 'DietPlan']
)
fig.update_layout(height=600)
fig.show()
```

```python
fig = px.box(
    workout_df, 
    x='ExercisePaln', 
    y='HeartRate', 
    title='Heart Rate by Exercise Plan',
    color='ExercisePaln',
    points='outliers'
)
fig.update_layout(showlegend=False)
fig.show()
```

```python
corr = workout_df.corr()
fig = px.imshow(
    corr,
    text_auto=True,
    title='Correlation Matrix - Workout Dataset',
    color_continuous_scale='RdBu',
    zmin=-1, zmax=1
)
fig.update_layout(height=500)
fig.show()
```

## 7. Cross-Dataset Comparisons

```python
age_data = []
for name, path in sources.items():
    df = pd.read_csv(path)
    if 'Age' in df.columns:
        temp = df[['Age']].copy()
        temp['Dataset'] = name
        age_data.append(temp)

age_combined = pd.concat(age_data, ignore_index=True)

fig = px.violin(
    age_combined, 
    x='Dataset', 
    y='Age', 
    title='Age Distribution Across Datasets',
    color='Dataset',
    box=True,
    points='outliers'
)
fig.update_layout(showlegend=False, xaxis_tickangle=15)
fig.show()
```

```python
bmi_data = []
for name, path in sources.items():
    df = pd.read_csv(path)
    if 'BMI' in df.columns:
        temp = df[['BMI']].copy()
        temp['Dataset'] = name
        bmi_data.append(temp)

bmi_combined = pd.concat(bmi_data, ignore_index=True)

fig = px.violin(
    bmi_combined, 
    x='Dataset', 
    y='BMI', 
    title='BMI Distribution Across Datasets',
    color='Dataset',
    box=True,
    points='outliers'
)
fig.update_layout(showlegend=False, xaxis_tickangle=15)
fig.show()
```

```python
fig = px.box(
    age_combined, 
    x='Dataset', 
    y='Age', 
    title='Age Distribution Comparison (Box Plot)',
    color='Dataset',
    points='outliers'
)
fig.update_layout(showlegend=False, xaxis_tickangle=15)
fig.show()
```

**Key Interactive Features:**

- **Reset**: Double-click to reset the view
- **Legend**: Click legend items to show/hide series
