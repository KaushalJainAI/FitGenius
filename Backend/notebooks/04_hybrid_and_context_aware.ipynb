# 04 - Hybrid Strategy
**Unit 4 & 6: Cascading Filters and Safety Constraints**

Implementing medical-aware rules to the similarity results to fulfill the 'Cascade' hybrid requirement.


```python
def cascade_filter(plans, health_record):
    """Hard-filter recommendations for diabetic users (Unit 4 Cascade).
    Checks Chronic_Disease field for diabetes-related conditions."""
    chronic_disease = str(health_record.get('Chronic_Disease', '')).lower()
    if 'diabetes' in chronic_disease or health_record.get('diabetes'):
        return [p for p in plans if 'Sugar' not in str(p)]
    return plans

def context_aware_adjustment(score, context):
    """Adjust recommendation scores based on wearable/contextual data."""
    sleep_quality = context.get('Sleep_Quality', 'Good')
    if sleep_quality == 'Poor':
        score *= 0.8  # Reduce high-intensity confidence
    daily_steps = context.get('Daily_Steps', 0)
    if daily_steps < 5000:
        score *= 0.9
    return score

print("✓ Cascade Hybrid logic ready for deployment.")
print("✓ Context-aware adjustments ready for deployment.")
```
